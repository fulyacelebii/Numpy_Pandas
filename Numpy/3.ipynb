### rand & randn & randint



### rand

```python
np.random.rand()
```

```python
np.random.rand(2,3)
```

### randn

**random.randn():** ortalama değeri 0 ve standart sapması 1 olan normal dağılıma sahip rastgele sayılar üretir.



```python
np.random.randn()
```

```python
np.random.randn(3,4)
```

```python
np.random.randn(300,400).mean() #ortalama sıfıra oldukça yakın.
```

```python
np.random.randn(300,400).std() #std 1'e oldukça yakın.
```

### Concatenation
Tek boyutlarda her zaman yan yana ekleme yapar.

- Satırlar için: axis = 0
- Sütunlar için: axis = 1 

```python
x = np.arange(1,7)
y = np.arange(7,13)
xy = np.concatenate((x,y)) #tek boyutların birleşimi yan yana olur.
xy
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
```

```python
ab = np.concatenate((a,b),  axis = 0) #default axis zaten 0
ab
```

```python
ab = np.concatenate((a,b),  axis = 1) #default axis zaten 0
ab
```

Peki ya farklı sayılarda matris olsaydı?

```python
c = np.array([[1,2],[3,4]])
d = np.array([[5,6],[7,8],[9,10]])
```

```python
cd = np.concatenate((c,d)) #Satırlar düzeyinde alt alta geleceği için bir sorun yaşanmadan çalıştı.
cd 
```

```python
cd = np.concatenate((c,d), axis = 1) #HATA çünkü yan yana geldiklerinde eksik var.
cd
```

### Split


**split():**, numpy dizisini belirtilen eksende eşit parçalara bölmek için kullanılır.

```python
array3 = np.arange(1,9)
array3
```

```python
np.split(array3,4) #array3'ü tam 4 parçaya bölecek
```

**array_split():** fonksiyonu, NumPy'da bir diziyi belirtilen sayıda eşit olmayan parçalara bölmek için kullanılır. Bu fonksiyon, diğer bölme fonksiyonlarından (hsplit, vsplit gibi) farklı olarak belirli bir eksende değil, kullanıcının belirttiği sayıda parçaya böler.

```python
np.array_split(array3,5) #tam bölmeyeceksek array_split kullanırız.
```

Ayrıca index girerek de bölme işlemi yaptırabiliriz.

```python
np.array_split(array3,[3]) #3. indexten böl dedik.
```

```python
np.array_split(array3,[3,6]) #iki farklı bölme de yapabiliriz
```
